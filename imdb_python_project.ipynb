```python
def rating(rating):
    if rating >= 7:
        return "excellent"
    elif rating >= 6:
        return "good"
    else:
        return "average"

```

```python
df['rating_name']=df['Rating'].apply(rating)
df.head()
```

```python
# find unique values from genre
```

```python
df['Genre']
```

```python
list1 = []
for value in df['Genre']:
    list1.append(value.split(','))
```

```python
one_d=[]
for item in list1 :
    for item1 in item :
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list =[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
